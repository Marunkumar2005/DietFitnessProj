```python
# Core
import os
import sys
import warnings
warnings.filterwarnings("ignore")

# Numerics & data
import numpy as np
import pandas as pd

# Modeling utilities
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Metrics
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    mean_absolute_error,
    r2_score
)

# Models (CPU-friendly; good for 8GB RAM)
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.neighbors import NearestNeighbors  # for similarity-based recommendations

# (Optional) Gradient boosting — use only if xgboost is installed
try:
    from xgboost import XGBClassifier, XGBRegressor
    XGB_AVAILABLE = True
except Exception:
    XGB_AVAILABLE = False

# Persistence
import joblib

# Plotting (for quick EDA later)
import matplotlib.pyplot as plt

# Reproducibility
RNG_SEED = 42
np.random.seed(RNG_SEED)

print("Imports OK. XGBoost available:", XGB_AVAILABLE)
```

```python
# Step 2: Load Dataset

# Example: assuming dataset is stored in 'fitness_diet_data.csv'
DATA_PATH = "data/diet_recommendations_dataset.csv"  

# Load CSV (or Excel fallback)
if DATA_PATH.endswith(".csv"):
    df = pd.read_csv(DATA_PATH)
elif DATA_PATH.endswith((".xls", ".xlsx")):
    df = pd.read_excel(DATA_PATH)
else:
    raise ValueError("Unsupported file format. Please use CSV or Excel.")

# Quick look at data
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())
```

```python
# Step 3: Preprocessing & Feature/Target setup

# Drop ID column (not useful for prediction)
df = df.drop(columns=["Patient_ID"])

# Define target
TARGET_COL = "Diet_Recommendation"
y = df[TARGET_COL]
X = df.drop(columns=[TARGET_COL])

print("Feature Shape:", X.shape)
print("Target Shape:", y.shape)

# Identify numeric vs categorical features
numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X.select_dtypes(exclude=[np.number]).columns.tolist()

print("\nNumeric Features:", numeric_features)
print("\nCategorical Features:", categorical_features)

# Handle missing values quickly (basic imputation)
# - fill numeric NaN with median
# - fill categorical NaN with mode
for col in numeric_features:
    X[col] = X[col].fillna(X[col].median())

for col in categorical_features:
    X[col] = X[col].fillna(X[col].mode()[0])

print("\nMissing values after imputation:\n", X.isnull().sum().sum())

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RNG_SEED, stratify=y
)

print("\nTrain Shape:", X_train.shape, "Test Shape:", X_test.shape)
```

```python
# Step 4: Encoding & Scaling

# Preprocessing for numeric features
numeric_transformer = StandardScaler()

# Preprocessing for categorical features (OneHotEncoding)
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

# Combine into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

print("Preprocessing pipeline created successfully.")
```

```python
# Step 5: Baseline Model with RandomForest

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# Build full pipeline: preprocessing + model
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(
        n_estimators=200,   # number of trees
        max_depth=None,     # let trees grow fully
        random_state=RNG_SEED,
        n_jobs=-1           # use all cores
    ))
])

# Train model
rf_pipeline.fit(X_train, y_train)

# Predictions
y_pred = rf_pipeline.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
import numpy as np

# Define parameter grid for RandomForest
param_dist = {
    "classifier__n_estimators": [100, 200, 300, 500],
    "classifier__max_depth": [None, 5, 10, 20, 30],
    "classifier__min_samples_split": [2, 5, 10],
    "classifier__min_samples_leaf": [1, 2, 4],
    "classifier__max_features": ["sqrt", "log2", None],
    "classifier__bootstrap": [True, False]
}

# Build pipeline with RandomForest
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=RNG_SEED, n_jobs=-1))
])

# RandomizedSearchCV for faster hyperparameter tuning
rf_search = RandomizedSearchCV(
    estimator=rf_pipeline,
    param_distributions=param_dist,
    n_iter=30,              # number of random combinations to try
    scoring="f1_weighted",  # optimize for overall balance
    cv=5,                   # 5-fold CV
    verbose=2,
    random_state=RNG_SEED,
    n_jobs=-1
)

# Train tuned model
rf_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", rf_search.best_params_)

# Evaluate on test set
best_rf = rf_search.best_estimator_
y_pred = best_rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# ===== Manual prediction helper (interactive version) =====
import joblib
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

MODEL_PATH = "random_forest_pipeline.pkl"

# Ask user for manual input
USER_INPUT = {
    "Age": int(input("Enter Age: ")),
    "Gender": input("Enter Gender (Male/Female): "),
    "Weight_kg": float(input("Enter Weight (kg): ")),
    "Height_cm": float(input("Enter Height (cm): ")),
    "BMI": float(input("Enter BMI: ")),
    "Disease_Type": input("Enter Disease Type (None/Diabetes/Hypertension/Obesity/etc.): "),
    "Severity": input("Enter Severity (Mild/Moderate/Severe): "),
    "Physical_Activity_Level": input("Enter Physical Activity Level (Sedentary/Moderate/High): "),
    "Daily_Caloric_Intake": float(input("Enter Daily Caloric Intake: ")),
    "Cholesterol_mg/dL": float(input("Enter Cholesterol (mg/dL): ")),
    "Blood_Pressure_mmHg": float(input("Enter Blood Pressure (mmHg): ")),
    "Glucose_mg/dL": float(input("Enter Glucose (mg/dL): ")),
    "Dietary_Restrictions": input("Enter Dietary Restrictions (None/Low_Sugar/Low_Sodium/etc.): "),
    "Allergies": input("Enter Allergies (None/Peanuts/Gluten/etc.): "),
    "Preferred_Cuisine": input("Enter Preferred Cuisine (Indian/Chinese/Mexican/Italian/etc.): "),
    "Weekly_Exercise_Hours": float(input("Enter Weekly Exercise Hours: ")),
    "Adherence_to_Diet_Plan": float(input("Enter Adherence to Diet Plan (0–100): ")),
    "Dietary_Nutrient_Imbalance_Score": float(input("Enter Nutrient Imbalance Score: "))
}


# Map diet -> fitness level + score
DIET_TO_FITNESS = {
    "Balanced":   ("Excellent", 5),
    "Low_Carb":   ("Medium", 3),
    "Low_Sodium": ("Low", 1)
}
NUM_TO_DIET_FALLBACK = {0: "Balanced", 1: "Low_Carb", 2: "Low_Sodium"}

# Load pipeline
pipeline = joblib.load(MODEL_PATH)

# Extract expected features from ColumnTransformer
num_cols, cat_cols = [], []
for _, step in pipeline.named_steps.items():
    if isinstance(step, ColumnTransformer):
        for _, transformer, cols in step.transformers_:
            if cols in ("drop", None):
                continue
            cols = list(cols) if isinstance(cols, (list, tuple, np.ndarray)) else [cols]
            tname = (transformer.__class__.__name__ if transformer else "").lower()
            if any(k in tname for k in ["standard", "minmax", "scale", "imputer"]):
                num_cols.extend(cols)
            else:
                cat_cols.extend(cols)

expected_cols = num_cols + cat_cols

# Convert input into DataFrame
row = {c: USER_INPUT.get(c, (0 if c in num_cols else "None")) for c in expected_cols}
df = pd.DataFrame([row], columns=expected_cols)

# Convert numerics
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# Map adherence Yes/No if needed
if "Adherence_to_Diet_Plan" in df.columns and df["Adherence_to_Diet_Plan"].dtype == object:
    df["Adherence_to_Diet_Plan"] = (
        df["Adherence_to_Diet_Plan"].astype(str).str.strip().str.lower()
          .map({"yes": 100, "no": 0}).fillna(0)
    )

# Predict
pred_raw = pipeline.predict(df)[0]

# Resolve label
clf = pipeline.named_steps.get("classifier", list(pipeline.named_steps.values())[-1])
classes = getattr(clf, "classes_", None)
if classes is not None and isinstance(classes[0], str):
    diet_label = pred_raw
else:
    diet_label = NUM_TO_DIET_FALLBACK.get(int(pred_raw), str(pred_raw))

fitness_label, fitness_score = DIET_TO_FITNESS.get(diet_label, ("Unknown", 0))

print("\n=== Prediction ===")
print("Diet Recommendation:", diet_label)
print("Fitness Level:", fitness_label)
print("Fitness Score (0-5):", fitness_score)
```

```python
# ===== Manual prediction helper (interactive version with diet tips) =====
import joblib
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

MODEL_PATH = "random_forest_pipeline.pkl"

# Ask user for manual input
USER_INPUT = {
    "Age": int(input("Enter Age: ")),
    "Gender": input("Enter Gender (Male/Female): "),
    "Weight_kg": float(input("Enter Weight (kg): ")),
    "Height_cm": float(input("Enter Height (cm): ")),
    "BMI": float(input("Enter BMI: ")),
    "Disease_Type": input("Enter Disease Type (None/Diabetes/Hypertension/Obesity): "),
    "Severity": input("Enter Severity (Mild/Moderate/Severe): "),
    "Physical_Activity_Level": input("Enter Physical Activity Level (Active/Sedentary/Moderate): "),
    "Daily_Caloric_Intake": float(input("Enter Daily Caloric Intake: ")),
    "Cholesterol_mg/dL": float(input("Enter Cholesterol (mg/dL): ")),
    "Blood_Pressure_mmHg": float(input("Enter Blood Pressure (mmHg): ")),
    "Glucose_mg/dL": float(input("Enter Glucose (mg/dL): ")),
    "Dietary_Restrictions": input("Enter Dietary Restrictions (None/Low_Sugar/Low_Sodium): "),
    "Allergies": input("Enter Allergies (None/Peanuts/Gluten): "),
    "Preferred_Cuisine": input("Enter Preferred Cuisine (Indian/Chinese/Mexican/Italian): "),
    "Weekly_Exercise_Hours": float(input("Enter Weekly Exercise Hours: ")),
    "Adherence_to_Diet_Plan": float(input("Enter Adherence to Diet Plan (0–100): ")),
    "Dietary_Nutrient_Imbalance_Score": float(input("Enter Nutrient Imbalance Score: "))
}

# Map diet -> fitness level + score
DIET_TO_FITNESS = {
    "Balanced":   ("Excellent", 5),
    "Low_Carb":   ("Medium", 3),
    "Low_Sodium": ("Low", 1)
}
NUM_TO_DIET_FALLBACK = {0: "Balanced", 1: "Low_Carb", 2: "Low_Sodium"}

# Extra: Diet food suggestions
DIET_FOOD_GUIDE = {
    "Balanced": {
        "Foods to Eat": ["Vegetables", "Fruits", "Whole grains", "Lean proteins"],
        "Foods to Avoid": ["Excess sugar", "Fried foods"]
    },
    "Low_Carb": {
        "Foods to Eat": ["Leafy greens", "Eggs", "Fish", "Nuts", "Avocado"],
        "Foods to Avoid": ["Bread", "Pasta", "Rice", "Sugary drinks"]
    },
    "Low_Sodium": {
        "Foods to Eat": ["Bananas", "Oats", "Leafy greens", "Fish", "Unsalted nuts"],
        "Foods to Avoid": ["Processed foods", "Pickles", "Chips", "Canned soups"]
    }
}

# Load pipeline
pipeline = joblib.load(MODEL_PATH)

# Extract expected features from ColumnTransformer
num_cols, cat_cols = [], []
for _, step in pipeline.named_steps.items():
    if isinstance(step, ColumnTransformer):
        for _, transformer, cols in step.transformers_:
            if cols in ("drop", None):
                continue
            cols = list(cols) if isinstance(cols, (list, tuple, np.ndarray)) else [cols]
            tname = (transformer.__class__.__name__ if transformer else "").lower()
            if any(k in tname for k in ["standard", "minmax", "scale", "imputer"]):
                num_cols.extend(cols)
            else:
                cat_cols.extend(cols)

expected_cols = num_cols + cat_cols

# Convert input into DataFrame
row = {c: USER_INPUT.get(c, (0 if c in num_cols else "None")) for c in expected_cols}
df = pd.DataFrame([row], columns=expected_cols)

# Convert numerics
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# Map adherence Yes/No if needed
if "Adherence_to_Diet_Plan" in df.columns and df["Adherence_to_Diet_Plan"].dtype == object:
    df["Adherence_to_Diet_Plan"] = (
        df["Adherence_to_Diet_Plan"].astype(str).str.strip().str.lower()
          .map({"yes": 100, "no": 0}).fillna(0)
    )

# Predict
pred_raw = pipeline.predict(df)[0]

# Resolve label
clf = pipeline.named_steps.get("classifier", list(pipeline.named_steps.values())[-1])
classes = getattr(clf, "classes_", None)
if classes is not None and isinstance(classes[0], str):
    diet_label = pred_raw
else:
    diet_label = NUM_TO_DIET_FALLBACK.get(int(pred_raw), str(pred_raw))

fitness_label, fitness_score = DIET_TO_FITNESS.get(diet_label, ("Unknown", 0))

print("\n=== Prediction ===")
print("Diet Recommendation:", diet_label)
print("Fitness Level:", fitness_label)
print("Fitness Score (0-5):", fitness_score)

# Add diet suggestions
if diet_label in DIET_FOOD_GUIDE:
    food_info = DIET_FOOD_GUIDE[diet_label]
    print("\nSuggested Foods to Eat:", ", ".join(food_info["Foods to Eat"]))
    print("Foods to Avoid:", ", ".join(food_info["Foods to Avoid"]))
```
